```python
# add_details函数
# 找出（基因名）以ERCC开头的的列，将这些列的索引作为新的一行
# 提取默认的计数矩阵（即原始RNA计数）
# 计算每个细胞的总计数，将其作为size factors保存，用于后续的归一化
# 归一化count值，进行arcsin-sqrt转换，合并基因名注释，更新行注释，返回增强版的sce
def add_details(adata,gene_names_df):
    # adata.var["is_ERCC"] = adata.var_names.str.startswith["ERCC"]  adata.var_names 是一个 Index 对象，不是 Series，它没有 .str 这个方法的直接支持。
    adata.var['is_ERCC'] = pd.Series(adata.var_names).str.startswith('ERCC').values

    
    # 计算每个细胞的 size factor（总计数）
    size_factors = adata.X.sum(axis=1).A1 if hasattr(adata.X, 'A1') else adata.X.sum(axis=1)
    print(size_factors.shape)
    adata.obs['size_factors'] = size_factors # 把这个向量作为每个细胞的size factor，保存作为细胞的元数据

    # 归一化 counts
    norm_counts = adata.X / size_factors[:, np.newaxis]
    adata.layers['norm_counts'] = norm_counts

    # arcsine square root transform
    phicounts = 2 * np.arcsin(np.sqrt(norm_counts)) # 先平方根，压缩大值差异，再平滑非线性变换，再放大变换结果
    adata.layers['phicounts'] = phicounts

    # 添加 gene annotation（gene_names_df 有一列 'GeneID' 对应 adata.var_names）
    row_data = pd.DataFrame({'GeneID': adata.var_names})
    row_data = row_data.merge(gene_names_df, on='GeneID', how='left')
    row_data.index = adata.var_names  # 确保 index 对齐
    adata.var = pd.concat([adata.var, row_data.drop(columns='GeneID')], axis=1)

    return adata
    
    
```

```python
ciona16_sce = add_details(ciona16_sce,gene_names)
ciona16MiSeq_sce = add_details(ciona16MiSeq_sce,gene_names)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
from matplotlib.colors import ListedColormap
from palettable.colorbrewer.sequential import GnBu_8
```

```python
# 去掉 ERCC spike-in 基因 和 全为零的基因，只保留有表达的真实基因。
# 1. 计算每个基因的总表达量（对每列求和）
gene_total_counts = ciona16_sce.X.sum(axis=0)

print(type(gene_total_counts))

# 如果是稀疏矩阵，加 .A1
if hasattr(gene_total_counts, 'A1'):
    gene_total_counts = gene_total_counts.A1
# 在单细胞分析中，adata.X 常常是稀疏矩阵（sparse matrix），为了节省内存。常见格式是 scipy.sparse.csr_matrix。 A1是一个快捷方式，意思是：把一个 1×N 或 N×1 的稀疏矩阵变成一个一维 numpy array。
# 因为对于稀疏矩阵，sum的结果是一个稀疏矩阵，不是一个numpy.array，而是scipy.sparse._matrix，不能像普通的一维数组那样直接用。

# 2. 找出总表达为 0 的基因
zero_genes = gene_total_counts == 0

# 3. 保留 非 ERCC 且表达不为 0 的基因
gene_filter = (~ciona16_sce.var['is_ERCC']) & (~zero_genes)

# 4. 子集化得到新的 AnnData（行是细胞，列是基因）
ssData = ciona16_sce[:, gene_filter].copy()
```

```python
phi_values = ssData.layers["phicounts"]  # shape: (cells, genes)
# 如果是稀疏矩阵，先转成密集的
if hasattr(phi_values, "toarray"):
    phi_values = phi_values.toarray()
cor_data = np.corrcoef(phi_values)
# GnBu_9.colors 是一个 RGB 列表，可以插值生成更多颜色
palette = sns.color_palette(GnBu_8.mpl_colors, n_colors=100)
```

```python
from sklearn.decomposition import PCA
```

```python
# 假设 phi_values 是 (genes × cells)，R 中你用的是 transpose，所以我们要转置
# phi_values = ciona16_sce.obsm['phicounts'].T  # shape: (n_cells, n_genes)

# PCA 分解
pca = PCA(n_components=2)
pca_result = pca.fit_transform(phi_values)

# 构建 DataFrame 用于绘图
pca_df = pd.DataFrame({
    "PC1": pca_result[:, 0],
    "PC2": pca_result[:, 1],
    "cell_type": ciona16_sce.obs["cell_type"].values,
    "Embryo": ciona16_sce.obs["Embryo"].values
})

# 创建调色板并映射颜色
embryo_labels = pca_df["Embryo"].unique()
palette = sns.color_palette("Set1", n_colors=len(embryo_labels))
embryo_color_map = {label: palette[i] for i, label in enumerate(embryo_labels)}

# 绘制 PCA 图
plt.figure(figsize=(10, 8))
for label in embryo_labels:
    subset = pca_df[pca_df["Embryo"] == label]
    plt.scatter(subset["PC1"], subset["PC2"], label=label, c=[embryo_color_map[label]]*len(subset), alpha=0.7)

# 添加标签（CellType）
for i in range(pca_df.shape[0]):
    plt.text(pca_df["PC1"][i], pca_df["PC2"][i], pca_df["cell_type"][i], fontsize=8, alpha=0.6)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of phiValues")
plt.legend(title="Embryo", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# 保存为 PDF
plt.savefig("images/pca.pdf")
```

```python
# ➤ HiSeq 处理
hi_detected = (ciona16_sce.X > 0).sum(axis=1)  # 每行是细胞，每个细胞检测到的基因数
if hasattr(hi_detected, "A1"):  # 若是稀疏矩阵
    hi_detected = hi_detected.A1

hi_df = pd.DataFrame({
    "detected": hi_detected,
    "embryo": ciona16_sce.obs["Embryo"].values,
    "cell": ciona16_sce.obs["cell_type"].values,
    "sequencer": "HiSeq"
})

# ➤ MiSeq 处理
mi_detected = (ciona16MiSeq_sce.X > 0).sum(axis=1)
if hasattr(mi_detected, "A1"):
    mi_detected = mi_detected.A1

mi_df = pd.DataFrame({
    "detected": mi_detected,
    "embryo": ciona16MiSeq_sce.obs["Embryo"].values,
    "cell": ciona16MiSeq_sce.obs["cell_type"].values,
    "sequencer": "MiSeq"
})
```

```python
combined_df = pd.concat([hi_df, mi_df], ignore_index=True)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=combined_df, x="embryo", y="detected", hue="sequencer")

plt.title("Gene Detection per Cell across Embryos (HiSeq vs MiSeq)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/genesDetectedBoxPlot.pdf")
```

```python
# 构建并合并 HiSeq 和 MiSeq 的检测数（每行对应一个细胞）
scatter_df = pd.DataFrame({
    "HiSeq": hi_df["detected"].values,
    "MiSeq": mi_df["detected"].values,
    "embryo": hi_df["embryo"].values  # 假设两个数据集 embryo 是一样的
})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=scatter_df, x="MiSeq", y="HiSeq", hue="embryo")

plt.title("Detected Genes per Cell: MiSeq vs HiSeq")
plt.xlabel("MiSeq detected genes")
plt.ylabel("HiSeq detected genes")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("images/genesDetectedScatter.pdf")
```
